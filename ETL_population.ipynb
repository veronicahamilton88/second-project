```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Set up the connection to the csv file
population = "./Resources/us-population.csv"
population_df = pd.read_csv(population)
population_df.head(10)
```

```python
# Isolated the row United States populations
population_transformed = population_df.loc[(population_df["Location"]=="United States")].copy()
# Dropped commas from the population data and re-cast the number as a float instead of a string
population_transformed["2000"] = population_transformed["2000"].str.replace(",","").astype(float)
population_transformed["2001"] = population_transformed["2001"].str.replace(",","").astype(float)
population_transformed["2002"] = population_transformed["2002"].str.replace(",","").astype(float)
population_transformed["2003"] = population_transformed["2003"].str.replace(",","").astype(float)
population_transformed["2004"] = population_transformed["2004"].str.replace(",","").astype(float)
population_transformed["2005"] = population_transformed["2005"].str.replace(",","").astype(float)
population_transformed["2006"] = population_transformed["2006"].str.replace(",","").astype(float)
population_transformed["2007"] = population_transformed["2007"].str.replace(",","").astype(float)
population_transformed["2008"] = population_transformed["2008"].str.replace(",","").astype(float)
population_transformed["2009"] = population_transformed["2009"].str.replace(",","").astype(float)
# Show the table 
population_transformed
```

```python
# Set up the engine connection
connection_string = "group3:Butler@localhost:5432/pollution_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Check the engine connection
engine.table_names()
```

```python
# Load the dataframe into our database
population_transformed.to_sql(name="population", con=engine, if_exists='append', index=True)
```
